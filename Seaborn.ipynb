```python
#Q1
'''
Scatter Plot (sns.scatterplot):
Use: Visualize the relationship between two continuous variables. It's useful for identifying patterns or correlations between variables.

Histogram (sns.histplot):
Use: Display the distribution of a single numerical variable. It shows the frequency or count of data points within certain ranges (bins).

Bar Plot (sns.barplot):
Use: Display categorical data with rectangular bars. It shows the relationship between a categorical variable and a numerical variable. Can be used to compare values across categories.

Box Plot (sns.boxplot):
Use: Summarize the distribution of numerical data through quartiles. It's useful for detecting outliers and comparing distributions across different categories.

Heatmap (sns.heatmap):
Use: Plot rectangular data as a color-encoded matrix. It's useful for visualizing correlations in a dataset, especially when dealing with large datasets and correlations between many variables.
```

```python
#Q2
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plotting the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title('FMRI Signal over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.grid(True)
plt.legend(title='Event')
plt.show()
```

```python
#Q3
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plotting box plots
plt.figure(figsize=(12, 6))

# Box plot for age across passenger classes
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Age Distribution across Passenger Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Box plot for fare across passenger classes
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Fare Distribution across Passenger Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
#Q4
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plotting histogram with hue for 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True, palette='viridis')
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.grid(True)
plt.legend(title='Cut')
plt.show()
```

```python
#Q5
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plotting pair plot with hue for 'species'
sns.pairplot(iris_data, hue='species', palette='viridis', markers=['o', 's', 'D'])
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
#Q6
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Flight Passengers')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
