```python
#Q2
'''
In Bokeh, glyphs are visual markers that represent data points in plots. 
They include geometric shapes (like circles, squares, triangles), lines, and other graphical elements that can be added to a plot to visually represent data. 
'''
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Output to notebook (for Jupyter notebooks) or use output_file() for saving to HTML
output_notebook()

# Create a figure object
p = figure(title='Bokeh Plot with Glyphs', x_axis_label='x-axis', y_axis_label='y-axis')

# Add circle glyph with x and y coordinates
p.circle(x, y, size=10, color='navy', alpha=0.5)

# Add a line glyph (connecting points) with x and y coordinates
p.line(x, y, line_width=2, color='green')

# Show the plot
show(p)
```

```python
#Q3
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Output to notebook (for Jupyter notebooks) or use output_file() for saving to HTML
output_notebook()

# Create a figure object
p = figure(title='Customized Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis',
           plot_width=800, plot_height=600,
           title_text_font='times', title_text_font_size='16pt',
           x_axis_label_text_font='times', x_axis_label_text_font_size='14pt',
           y_axis_label_text_font='times', y_axis_label_text_font_size='14pt',
           major_label_text_font='helvetica', major_label_text_font_size='12pt',
           major_tick_line_color='navy', minor_tick_line_color='gray',
           background_fill_color='beige', border_fill_color='whitesmoke',
           grid_line_color='gray', grid_line_alpha=0.3, grid_line_dash='dashed')

# Add circle glyph with x and y coordinates
p.circle(x, y, size=10, color='navy', alpha=0.5, legend_label='Data Points')

# Customize legend
p.legend.location = 'top_left'
p.legend.title = 'Legend Title'
p.legend.background_fill_color = 'lightgray'

# Show the plot
show(p)
```

```python
#Q4
'''
A Bokeh server is a Python server application that allows you to build and deploy interactive Bokeh plots and applications. 
'''
# app.py
from bokeh.plotting import figure, curdoc
from bokeh.models import Slider
from bokeh.layouts import column
import numpy as np

# Create initial data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure
plot = figure(title="Interactive Sine Wave")
line = plot.line(x, y, line_width=2)

# Define a callback function to update the plot
def update(attr, old, new):
    freq = slider.value
    y = np.sin(freq * x)
    line.data_source.data['y'] = y

# Create a slider widget
slider = Slider(start=1, end=10, value=1, step=0.1, title="Frequency")
slider.on_change('value', update)

# Create a layout
layout = column(slider, plot)

# Add layout to the current document (curdoc())
curdoc().add_root(layout)

```
