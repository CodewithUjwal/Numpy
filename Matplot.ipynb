```python
#Q1
'''
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python
Why is Matplotlib used?

Data Visualization: To create various types of plots and charts to visualize data.
Customization: Offers extensive customization options for every aspect of the plots.
Integration: Works well with NumPy, pandas, and other scientific computing libraries.
Interactivity: Can create interactive plots for data exploration.
Publication Quality: Generates high-quality graphs suitable for publication.

Five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: Used to display information as a series of data points connected by straight line segments.
Bar Plot: Represents categorical data with rectangular bars with lengths proportional to the values they represent.
Histogram: An approximate representation of the distribution of numerical data.
Scatter Plot: Uses Cartesian coordinates to display values for typically two variables for a set of data.
Pie Chart: A circular statistical graphic, which is divided into slices to illustrate numerical proportion.
```

```python
#Q2
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.7)  # Plotting the scatter plot
plt.title('Scatter Plot of X vs Y')  # Adding title
plt.xlabel('X')  # Adding label for x-axis
plt.ylabel('Y')  # Adding label for y-axis
plt.grid(True)  # Adding grid for better readability
plt.show()
```

```python
#Q3
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.figure(figsize=(12, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, marker='o', color='b')
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, marker='s', color='g')
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, marker='^', color='r')
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, marker='*', color='purple')
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```
